```python
class Matrix:
    def get_dimensions(self):
        rows = len(self.matrix)
        columns = len(self.matrix[0]) if rows > 0 else 0
        return rows, columns
    #классический init, чтобы вводить)
    def __init__(self, matrix):
        self.matrix = matrix
    #Здесь операция сложения
    def __add__(self, other):
        #создаем пустой список, где будет результат
        result = []
        #isinstance сравнивает other, является ли он интом или флотом
        if isinstance(other, float) or isinstance(other, int):
            #Если азер (байджан) является флотом или интом - действуем по логике суммы матрицы  и числа
            #цикл на длину матрицы
            for l in range(len(self.matrix)):
                #пустой список для новой матрицы
                p = []
                #проходим по колоннам, матрих 0 это первый элемент
                for col in range(len(self.matrix[0])):
                    #аппендируем (добавляем сумму матрицы и числа в конец списка p)
                    p.append(self.matrix[l][col] + other)
                result.append(p)
            #Если азер не флот и не инт, то действуем по логике суммы двух матриц
        else:
            for l in range(len(self.matrix)):
                p = []
                for col in range(len(self.matrix[0])):
                    p.append(self.matrix[l][col] + other.matrix[l][col])
                result.append(p)
        return Matrix(result)

    #Тут натуральный ctrl c + ctrl v
    def __sub__(self, other):
        #создаем пустой список, где будет результат
        result = []
        #isinstance сравнивает other, является ли он интом или флотом
        if isinstance(other, float) or isinstance(other, int):
            #Если азер (байджан) является флотом или интом - действуем по логике суммы матрицы  и числа
            #цикл на длину матрицы
            for l in range(len(self.matrix)):
                #пустой список для новой матрицы
                p = []
                #проходим по колоннам, матрих 0 это первый элемент
                for col in range(len(self.matrix[0])):
                    #аппендируем (добавляем сумму матрицы и числа в конец списка p)
                    p.append(self.matrix[l][col] - other)
                result.append(p)
            #Если азер не флот и не инт, то действуем по логике суммы двух матриц
        else:
            for l in range(len(self.matrix)):
                p = []
                for col in range(len(self.matrix[0])):
                    p.append(self.matrix[l][col] - other.matrix[l][col])
                result.append(p)
        return Matrix(result)
    
    #Да и тут впринципе
    def __mul__(self, other):
        #создаем пустой список, где будет результат
        result = []
        #isinstance сравнивает other, является ли он интом или флотом
        if isinstance(other, float) or isinstance(other, int):
            #Если азер (байджан) является флотом или интом - действуем по логике суммы матрицы  и числа
            #цикл на длину матрицы
            for l in range(len(self.matrix)):
                #пустой список для новой матрицы
                p = []
                #проходим по колоннам, матрих 0 это первый элемент
                for col in range(len(self.matrix[0])):
                    #аппендируем (добавляем сумму матрицы и числа в конец списка p)
                    p.append(self.matrix[l][col] * other)
                result.append(p)
            #Если азер не флот и не инт, то действуем по логике суммы двух матриц
        else:
            for l in range(len(self.matrix)):
                p = []
                for col in range(len(self.matrix[0])):
                    p.append(self.matrix[l][col] * other.matrix[l][col])
                result.append(p)
        return Matrix(result)

    #Да и здесь тоже
    def __truediv__(self, other):
        #создаем пустой список, где будет результат
        result = []
        #isinstance сравнивает other, является ли он интом или флотом
        if isinstance(other, float) or isinstance(other, int):
            #Если азер (байджан) является флотом или интом - действуем по логике суммы матрицы  и числа
            #цикл на длину матрицы
            for l in range(len(self.matrix)):
                #пустой список для новой матрицы
                p = []
                #проходим по колоннам, матрих 0 это первый элемент
                for col in range(len(self.matrix[0])):
                    #аппендируем (добавляем сумму матрицы и числа в конец списка p)
                    p.append(self.matrix[l][col] / other)
                result.append(p)
            #Если азер не флот и не инт, то действуем по логике суммы двух матриц
        else:
            for l in range(len(self.matrix)):
                p = []
                for col in range(len(self.matrix[0])):
                    p.append(self.matrix[l][col] / other.matrix[l][col])
                result.append(p)
        return Matrix(result)
    
    #А вот матьмул по другому делается
    def __matmul__(self, other):
        #не делаем isinstance, т.к. операции могут проходить только между двумя матрицами
        #cначала проверяем, совпадает ли длина матрицы other (количество столбцов) с шириной матрицы self (количество строк). Если размеры матриц не совпадают, то функция вызывает ошибку
        if len(other.matrix) != len(self.matrix[0]):
            raise ValueError("Размеры матриц должны быть одинаковыми")

        #здесь все так же наш результат
        result = []
        #двойной цикл for для умножения элементов матриц self и other
        for i in range(len(self.matrix)):
            s = []
            for j in range(len(other.matrix[0])):
                p = 0
            #вычисляем сумму произведений элементов строки self и столбца other и добавляет результат к p
                for k in range(len(self.matrix[0])):
                    p += self.matrix[i][k] * other.matrix[k][j]
                s.append(p)
            result.append(s)

        return Matrix(result)
    
    def T(self ):
        rows = len(self)
        columns = len(self.matrix[0]) if rows > 0 else 0
        t = []
        for s in range(columns):
            row = []
            for i in range(rows):
                row.append(self.matrix[i][s])
            t.append(row)
        return Matrix(t)
    
    def __len__(self):
        return len(self.matrix)
    
    def __str__(self):
        return str(self.matrix)
    
    def get_vectors(self):
        if self.get_dimensions()[0] == 4 and self.get_dimensions()[1] == 4:
            v1, v2, v3, v4 = ([] for i in range(4))
            for i in self.matrix:
                v1.append(i[0])
                v2.append(i[1])
                v3.append(i[2])
                v4.append(i[3])
            return [v1,], [v2,], [v3,], [v4,]
        else:
            raise AttributeError("Only works with 4x4 matrixes!")
        
    @staticmethod
    def vector_multiply(vec1, vec2):
        res = 0
        for i in range(len(vec2[0])):
            res += vec1[0][i] * vec2[0][i]
        return res

    def basises(self):
        if self.get_dimensions()[0] == 4 and self.get_dimensions()[1] == 4:
            v1,v2,v3,v4 = self.get_vectors()
            f1 = v1

            f2 = (Matrix(v2) - (Matrix(f1)*(Matrix.vector_multiply(v2,f1) / Matrix.vector_multiply(f1,f1)))).matrix

            f3 = (Matrix(v3) - (Matrix(f1)*(Matrix.vector_multiply(v3,f1) / Matrix.vector_multiply(f1,f1))) -
                               (Matrix(f2)*(Matrix.vector_multiply(v3,f2) / Matrix.vector_multiply(f2,f2)))).matrix
            
            f4 = (Matrix(v4) - (Matrix(f1)*(Matrix.vector_multiply(v4,f1) / Matrix.vector_multiply(f1,f1))) -
                               (Matrix(f2)*(Matrix.vector_multiply(v4,f2) / Matrix.vector_multiply(f2,f2))) -
                               (Matrix(f3)*(Matrix.vector_multiply(v4,f3) / Matrix.vector_multiply(f3,f3)))).matrix
            return Matrix(f1), Matrix(f2), Matrix(f3), Matrix(f4)
        else:
            raise AttributeError("err")
    
    def orthogonals(self):
        if self.get_dimensions()[0] == 4 and self.get_dimensions()[1] == 4:
            res = []
            for f in self.basises():
                res.append((f/Matrix.norm(f)).matrix)
            return(res)
        else:
            raise AttributeError("err")
        #smth
    @staticmethod
    def norm(vec):
        print(vec.matrix)
        result = vec.matrix[0][0]**2 + vec.matrix[0][1]**2 + vec.matrix[0][2]**2 + vec.matrix[0][3]**2
        return result**0.5
```
