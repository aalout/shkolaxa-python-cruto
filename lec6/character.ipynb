```python
class MixinAbilityUsage:
    def rot(self):
        print('Ability')

class MixinSwimable:
    def swim(self):
        print('Холодец)')

class MixinTalkable:
    def make_noise(self):
        print('Фарш не провернуть назад... И мясо из котлет не восстановишь...')
        
class MixinFall:
    def fall(self):
        print('Пол проломился')
        
class MixinBreakable:
    def breakable(self):
        print('Ouch')
        
class MixinWalkable:
    def walk(self):
        print('Перекатываюсь')
```

```python
class Characters:
    def init(
        self,
        width_cm: float,
        height_cm: float
    ):
        self.width_cm: float = width_cm
        self.height_cm: float = height_cm
        
class Playable(MixinAbilityUsage, MixinWalkable, MixinTalkable):
    pass
        
class Toy(MixinBreakable, MixinSwimable, MixinFall):
    pass
        
class Pudge(Characters, Playable):
    width_cm: float = 100000
    hook_length_cm: float = 32000
    passive_stacks: float = 10
    
    def init(
        self,
        width_cm: float = 100000,
        hook_length_cm: float = 32000,
        passive_stacks: float = 10
    ):
        self.width_cm: float = width_cm
        self.hook_length: float = hook_length_cm
        self.passive_stacks: float = passive_stacks
        print('True Pudge Initialized')
        
    def hook():
        print('Падж тебя приголубит)')
        
class ToyPudge(Characters, Toy):
    stinky: bool = False
    size_cm3: float = 12
    width_cm: float = 8
    
    def init( 
        self,
        stinky: bool = False,
        size_cm3: float = 12,
        width_cm: float = 8
    ):
        self.stinky: bool  = stinky
        self.size_cm3: float = size_cm3
        self.width_cm: float = width_cm
        print('Toy')

    def stand(self):
        print('прицеливается хоком)')
        
    def head_turn(self):
        print('Поломался')
        
class Shadowfiend(Characters, Playable):
    shadowraze_counter: int = 3
    height_cm: float = 210
    
    def init(
        self,
        shadowraze_counter: int = 3,
        height_cm: float = 210
    ):
        self.shadowraze_counter: int = shadowraze_counter
        self.height_cm: float = height_cm
        
    def shadowraze():
        print('Coil, coil, coil')
        
class ToyShadowfiend(Characters, Toy):
    size_podstavka_cm2: float = 10
    height_cm: float = 15
    
    def init(
        self,
        size_podstavka_cm2: float = 10,
        height_cm: float = 15
    ):
        self.size_podstavka_cm2: float = size_podstavka_cm2
        self.height_cm: float = height_cm
        
    def podstavka_break():
        print('shadowfiend упал((')
        
class AntiMage(Characters, Playable):
    creep: bool = True
    farm: bool = False
    
    def init(
        self,
        creep: bool = True,
        farm: bool = False  
    ):
        self.creep: bool = creep
        self.farm: bool = farm
        
    def blink():
        print('Пока')
        
class ToyAntiMage(Characters, Toy):
    height_cm: float = 10
    blades_count: float = 2
    
    def init(
        self,
        height_cm: float = 10,
        blades_count: float = 2
    ):
        self.height_cm: float = 10
        self.blades_count: float = 2
        
    def nothing():
        print('nothing happened./,.')
    
```
